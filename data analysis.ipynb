```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fast_food_data = pd.read_csv ('FastFoodRestaurants.csv')
fast_food_data
```

```python
data_set = fast_food_data[["address", "city","country","latitude","longitude","name","postalCode","province"]]
data_set
```

```python
data_set["name"].value_counts()
```

```python
# check about datatypes and nulls
data_set.info()
```

```python
#display of the density of fast food restaurant in all the  provinces across the US

plt.figure(figsize=(20, 5))
sns.histplot(data_set['province'], kde=True, bins=15)
```

```python
# # top 10 province with fast food restaurants provinces
df = data_set["province"].value_counts()[:10]
df
```

```python
plt.figure(figsize=(20, 5))
df.plot(kind='bar', color='purple')
plt.title('Number of fast-food restaurant per Province', fontsize=14)
plt.xlabel('Province')
plt.ylabel('Frequency')
plt.show()
```

```python

df = data_set.groupby(["name", "postalCode"]).size().nlargest()
df
```

```python
#top 10 zip code with detailed restaurants occupation

df= df.unstack()
top_postal_codes = df.sum(axis=0).nlargest(10).index
top_counts = df[top_postal_codes]
top_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('fastfood restaurants')
plt.ylabel('Count')
```
